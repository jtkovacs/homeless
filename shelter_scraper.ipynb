```python
from bs4 import BeautifulSoup
import urllib.request
import csv
```

```python
HSD = urllib.request.urlopen('http://www.homelessshelterdirectory.org/cgi-bin/id/city.cgi?city=Seattle&state=WA').read()
SL = urllib.request.urlopen('http://www.shelterlistings.org/city/seattle-wa.html').read()
```

```python
HSD_soup = BeautifulSoup(HSD, "html.parser")
SL_soup = BeautifulSoup(SL, "html.parser")
#print(HSD_soup.prettify()[0:2000])
```

```python
HSD_divs = HSD_soup.find_all("div", class_="item_content")
HSD_shelters = list()
for em in HSD_divs:
    try:
        HSD_shelters.append(["HSD",em.a.get_text()])
    except:
        pass
HSD_shelters_final = HSD_shelters[0:-18]
HSD_shelters_final 
```

```python
SL_divs = SL_soup.find_all("td")
SL_shelters = list()
for em in SL_divs:
    try:
        tx = em.a.get_text()
        if len(tx) > 1: 
            SL_shelters.append(["SL", tx])
    except:
        pass
SL_shelters
```

```python
all_shelters = HSD_shelters_final + SL_shelters
```

```python
all_shelters.sort()
```

```python
all_shelters
```

```python
filename = 'resources_shelters_raw.csv'
with open(filename, 'w', newline='') as tsvfile:
    my_writer = csv.writer(tsvfile, delimiter='\t', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for em in all_shelters:
        my_writer.writerow(em)
```
